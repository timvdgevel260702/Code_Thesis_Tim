```python
import pandas as pd

dom_df = pd.read_csv("dom_hierarchy_features_v3.csv")
css_df = pd.read_csv("css_regularity_features_3130.csv")
ad_df  = pd.read_csv("ad_density_features_3130_v3.csv")

dom_df = dom_df.sort_values("url").reset_index(drop=True)
css_df = css_df.sort_values("url").reset_index(drop=True)
ad_df  = ad_df.sort_values("url").reset_index(drop=True)

id_cols = [
    "url",
    "mbfc-credibility-rating",
    "target",
    "html_filename",
    "css_filename"
]

dom_feature_cols = [
    "dom_max_depth",
    "dom_avg_branching",
    "dom_branching_std",      # Now capped
    "json_ld_present",
    "social_link_ratio",
    "semantic_tag_ratio",
    "nav_max_list_depth",
    "text_to_tag_ratio",
    "interactive_ratio",
    "external_link_ratio",
    "internal_to_external_link_ratio", # Now log-transformed
    "nav_link_ratio",
    "empty_link_ratio",
    "heading_density",
    "paragraph_density",
    "meta_description_present"
]

css_feature_cols = [
    "css_class_token_entropy",
    "css_reuse_factor",
    "structured_naming_ratio",
    "inline_style_density",
    "css_var_count",
]

ad_meta_cols = [
    "country",
    "media-type"
]

ad_feature_cols = [
    "iframe_ratio",
    "adtech_script_ratio",
    "explicit_ad_slot_count",
    "header_bidding_present",
    "script_tag_ratio",
    "third_party_script_ratio",
    "ad_keyword_present",
]

dom_keep = dom_df[id_cols + dom_feature_cols].copy()
css_keep = css_df[["url"] + css_feature_cols].copy()
ad_keep  = ad_df[["url"] + ad_meta_cols + ad_feature_cols].copy()

print("Before dedup:")
print("DOM:", dom_keep.shape, "| duplicate urls:", dom_keep["url"].duplicated().sum())
print("CSS:", css_keep.shape, "| duplicate urls:", css_keep["url"].duplicated().sum())
print("AD :", ad_keep.shape,  "| duplicate urls:", ad_keep["url"].duplicated().sum())

dom_keep = dom_keep.drop_duplicates(subset="url")
css_keep = css_keep.drop_duplicates(subset="url")
ad_keep  = ad_keep.drop_duplicates(subset="url")

print("\nAfter dedup:")
print("DOM:", dom_keep.shape)
print("CSS:", css_keep.shape)
print("AD :", ad_keep.shape)

merged_df = dom_keep.merge(css_keep, on="url", how="inner")
merged_df = merged_df.merge(ad_keep, on="url", how="inner")

print("\nMerged shape:", merged_df.shape)
print("Duplicate urls in merged:", merged_df["url"].duplicated().sum())
merged_df.head(10)
```

```python
merged_df.to_csv("scs_feature_table_final_3044.csv", index=False)
print("Saved: scs_feature_table_final_3044.csv")
```

```python
import pandas as pd
import numpy as np

# =========================
# ADD NEW CSS FEATURES TO FINAL MERGED DATASET
# =========================

# 1. load new CSS feature file
css_df = pd.read_csv("css_new_features_results.csv")

new_css_cols = [
    "css_selector_count",
    "boilerplate_selector_share",
    "icon_selector_share",
    "custom_dominant_prefix_share",
    "custom_prefix_family_count",
    "custom_bem_share",
    "media_query_density",
    "html_css_class_coverage",
    "unused_css_class_ratio",
]

css_subset = css_df[["url", "css_filename"] + new_css_cols].drop_duplicates(
    subset=["url", "css_filename"]
)

# 2. automatically find your final dataframe
candidate_names = ["final_df", "result_df", "merged_df", "analysis_df", "df"]
target_df_name = None
target_df = None

for name in candidate_names:
    if name in globals():
        obj = globals()[name]
        if isinstance(obj, pd.DataFrame) and {"url", "css_filename"}.issubset(obj.columns):
            target_df_name = name
            target_df = obj
            break

if target_df is None:
    raise ValueError(
        "Could not find the final dataframe automatically. "
        "Make sure your final dataframe exists and contains 'url' and 'css_filename'."
    )

print(f"Using dataframe: {target_df_name}")
print("Shape before merge:", target_df.shape)

# 3. merge new CSS features
target_df = target_df.merge(
    css_subset,
    on=["url", "css_filename"],
    how="left"
)

# 4. fill missing values in new CSS features
target_df[new_css_cols] = target_df[new_css_cols].replace([np.inf, -np.inf], np.nan).fillna(0)

# 5. drop old weak CSS features if they exist
old_css_drop_cols = [
    "structured_naming_ratio",
    "css_var_count",
]

target_df = target_df.drop(columns=[c for c in old_css_drop_cols if c in target_df.columns])

print("Shape after merge:", target_df.shape)

print("\nMissing values in new CSS columns:")
print(target_df[new_css_cols].isna().sum().to_string())

print("\nPreview of new CSS columns:")
display(target_df[["url", "css_filename"] + new_css_cols].head())

# 6. save updated merged file
target_df.to_csv("scs_feature_table_final_3044.csv", index=False)
print("\nSaved updated file: scs_feature_table_final_3044.csv")



# 7. write back into globals so later cells use the updated dataframe too
globals()[target_df_name] = target_df
```

```python
# Check duplicate URLs after merging feature tables

URL_COL = "url"   # change if your column is named differently

# Total duplicate rows, excluding the first occurrence
n_duplicate_rows = merged_df.duplicated(subset=[URL_COL]).sum()

# Number of unique URLs that appear more than once
n_duplicate_urls = merged_df[URL_COL].value_counts().gt(1).sum()

print("Duplicate rows:", n_duplicate_rows)
print("Duplicate unique URLs:", n_duplicate_urls)

# Show duplicated URLs and how often they occur
duplicate_url_counts = (
    merged_df[URL_COL]
    .value_counts()
    .loc[lambda x: x > 1]
    .reset_index()
)

duplicate_url_counts.columns = [URL_COL, "count"]

display(duplicate_url_counts)
```
